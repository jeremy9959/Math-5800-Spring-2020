```python
import networkx as nx
from itertools import product
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.DiGraph()
x_nodes = sorted(['x_'+str(i) for i in range(10)])
z_nodes = ['z_'+str(j) for j in range(1)]
w_nodes = ['output']
```

```python
G.add_edges_from(product(x_nodes,z_nodes))
G.add_edges_from(product(z_nodes,w_nodes))
```

```python
G.nodes
```

```python
plt.figure(figsize=(10,10))
pos = nx.drawing.layout.bipartite_layout(G,nodes=x_nodes)
pos['z_0']=np.array([1,0])
pos['output']=np.array([2,0])
colors, sizes=[],[]
for node in G.nodes:
    if node in x_nodes:
        colors.append('lightblue')
        sizes.append(500)
    elif node in z_nodes:
        colors.append('green')
        sizes.append(1000)
    else:
        colors.append('red')
        sizes.append(500)
ax=nx.nx_pylab.draw_networkx(G,pos=pos,node_color=colors,node_size=sizes)
_=plt.axis('off')
_=plt.suptitle("A Neuron",size=30,y=.9)
```

```python
G = nx.DiGraph()
x_nodes = sorted(['x_'+str(i) for i in range(10)])
z_nodes = ['z_'+str(j) for j in range(6)]
G.add_edges_from(product(x_nodes,z_nodes))
```

```python
plt.figure(figsize=(10,10))
pos = nx.drawing.layout.bipartite_layout(G,nodes=x_nodes)
colors=[]
for node in G.nodes:
    if node in x_nodes:
        colors.append('lightblue')
        sizes.append(500)
    elif node in z_nodes:
        colors.append('green')
        sizes.append(500)
    
ax=nx.nx_pylab.draw_networkx(G,pos=pos,node_color=colors)
_=plt.axis('off')
_=plt.suptitle("A Layer",size=30,y=.9)
```
