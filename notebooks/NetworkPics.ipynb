```python
import bokeh
from bokeh.io import show, output_notebook
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, Title, NormalHead
from bokeh.models.ranges import Range1d
from bokeh.models.annotations import LabelSet, Arrow
from bokeh.models.glyphs import Text
import pandas as pd
import numpy as np
from itertools import product
```

```python
output_notebook()
```

```python
f = figure(height=500,width=500, x_range=Range1d(-1,1),y_range=Range1d(-1,1),title='A Three Layer Network - Backward Propagation',toolbar_location=None)
f.xgrid.visible=False
f.ygrid.visible=False
f.xaxis.visible=False
f.yaxis.visible=False
f.outline_line_alpha = 0.0
```

```python

f = figure(height=500,width=500, x_range=Range1d(-1,1),y_range=Range1d(-1,1),title='A Three Layer Network - Backward Propagation',toolbar_location=None)
f.xgrid.visible=False
f.ygrid.visible=False
f.xaxis.visible=False
f.yaxis.visible=False
f.outline_line_alpha = 0.0
x = [-.75]*10
y = np.linspace(-.75,.75,10)
x1 = [0]*3
y1 = np.linspace(-.4,.4,3)
xsa = [[x[i],x1[j]] for i,j in product(range(1,10),range(3))]
ysa = [[y[i],y1[j]] for i,j in product(range(1,10),range(3))]

xsb = [[x[i],x1[j]] for i,j in product(range(0,1),range(3))]
ysb = [[y[i],y1[j]] for i,j in product(range(0,1),range(3))]
f.multi_line(xs=xsa,ys=ysa,line_color='gray',line_alpha=.2)
f.multi_line(xs=xsb,ys=ysb)

x2 = [.75]*5
y2 = np.linspace(-.6,.6,5)
xs = [[x1[i],x2[j]] for i,j in product(range(3),range(5))]
ys = [[y1[i],y2[j]] for i,j in product(range(3),range(5))]

f.multi_line(xs=xs,ys=ys,line_color='gray',line_alpha=0.2)
labels = ColumnDataSource({'x':[-.75,0,.75],'y':[.8,.8,.8],'name':['Input','Hidden','Output']})
node_labels = ColumnDataSource({'x':[-.75]*10,'y':np.linspace(-.75,.75,10),'name':['x_{}'.format(i) for i in range(10)]})
f.circle(x=x,y=y,radius=.02,fill_color='white',line_width=2)
f.circle(x=x1,y=y1,radius=.02,fill_color='white',line_width=2)
f.circle(x=x2,y=y2,radius=.02,fill_color='white',line_width=2)
f.add_layout(LabelSet(x='x',y='y',text='name',source=labels,x_offset=-25))
f.add_layout(LabelSet(x='x',y='y',text='name',source=node_labels,x_offset=-25,text_font_size='8pt'))
show(f)
```

```python

f = figure(height=500,width=500, x_range=Range1d(-1,1),y_range=Range1d(-1,1),title='A Three Layer Network - Forward Propagation',toolbar_location=None)
f.xgrid.visible=False
f.ygrid.visible=False
f.xaxis.visible=False
f.yaxis.visible=False
f.outline_line_alpha = 0.0
x = [-.75]*10
y = np.linspace(-.75,.75,10)
x1 = [0]*3
y1 = np.linspace(-.4,.4,3)
xsa = [[x[i],x1[j]] for i,j in product(range(0,10),range(1,3))]
ysa = [[y[i],y1[j]] for i,j in product(range(0,10),range(1,3))]

xsb = [[x[i],x1[j]] for i,j in product(range(0,10),range(0,1))]
ysb = [[y[i],y1[j]] for i,j in product(range(0,10),range(0,1))]
f.multi_line(xs=xsa,ys=ysa,line_color='gray',line_alpha=.2)
f.multi_line(xs=xsb,ys=ysb)

x2 = [.75]*5
y2 = np.linspace(-.6,.6,5)
xs = [[x1[i],x2[j]] for i,j in product(range(3),range(5))]
ys = [[y1[i],y2[j]] for i,j in product(range(3),range(5))]

f.multi_line(xs=xs,ys=ys,line_color='gray',line_alpha=.2)
labels = ColumnDataSource({'x':[-.75,0,.75],'y':[.8,.8,.8],'name':['Input','Hidden','Output']})
node_labels = ColumnDataSource({'x':[-.75]*10,'y':np.linspace(-.75,.75,10),'name':['x0_{}'.format(i) for i in range(10)]})
node_labels2 = ColumnDataSource({'x':[0]*3,'y':np.linspace(-.4,.4,3),'name':['x1_{}'.format(i) for i in range(3)]})

f.circle(x=x,y=y,radius=.02,fill_color='white',line_width=2)
f.circle(x=x1,y=y1,radius=.02,fill_color='white',line_width=2)
f.circle(x=x2,y=y2,radius=.02,fill_color='white',line_width=2)
f.add_layout(LabelSet(x='x',y='y',text='name',source=labels,x_offset=-25,y_offset=10))
f.add_layout(LabelSet(x='x',y='y',text='name',source=node_labels,x_offset=-30,text_font_size='8pt'))
f.add_layout(LabelSet(x='x',y='y',text='name',source=node_labels2,x_offset=-15,y_offset=20,text_font_size='8pt'))

show(f)
```

```python
neuron  = figure(title="A Neuron",width=500,height=500,x_range=Range1d(-1,1),y_range=Range1d(-1,1),toolbar_location=None)
neuron.xgrid.visible=False
neuron.ygrid.visible=False
neuron.xaxis.visible=False
neuron.yaxis.visible=False
neuron.outline_line_alpha = 0.0
xs = [[-.8,0]]*5
ys = [[y,0] for y in np.linspace(-.8,.8,5)]
neuron.multi_line(xs=xs,ys=ys)
neuron.line(x=[0,1],y=[0,0])
neuron.circle(x=[0],y=[0],radius=.1,fill_color='lightgray')
neuron.text(x=[.15],y=[0],text=['new activation=sigma(sum of inputs)'],text_font_size='9pt')
neuron.text(x=[-.8],y=[.8],text=['prior activation'],text_font_size='9pt')
neuron.text(x=[-.5],y=[.5],text=['weight'],text_font_size='9pt')
neuron.text(x=[-.2],y=[.2],text=['input = weight * prior activation'],text_font_size='9pt')
neuron.add_layout(Arrow(end=NormalHead(size=6,line_color='green',fill_color='green'),x_start=-.8,x_end=-.075,y_start=.8,y_end=.07,line_color='green'))
neuron.add_layout(Arrow(end=NormalHead(size=6),x_start=.9,x_end=1.0,y_start=0,y_end=0))
show(neuron)
```
